```python
from sklearn.linear_model import LinearRegression
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("expense_tracker").getOrCreate()

total_monthly = spark.table("workspace.expense_analytics_gold.gold_monthly_expense")
total_monthly.show()

pdf = total_monthly.toPandas()
```

```python
X = pdf[["month"]]
y = pdf["total_expense"]

model = LinearRegression()
model.fit(X, y)

predict_next_month = model.predict([[pdf["month"].max() + 1]])
print(predict_next_month)
```

```python
average_spend = pdf['total_expense'].mean()

if predict_next_month[0] > average_spend:
    print("❗ Overspending Alert: Next month spending may increase!")
else:
    print("✅ Spending is under control")
```
